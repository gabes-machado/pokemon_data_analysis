```python
# Importação de bibliotecas de análise e listagem de base de dados inseridas:
import matplotlib.pyplot as plt # plotting graphs
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import statsmodels.api as sm #statistical modelling
import math #mathematical functions
```

```python
# Create a dataframe from the csv file
df = pd.read_csv('..\poke_data\data\processed\pokemon_ready.csv')
display(df)
```

## Perguntas-problema:

>**Existe correlação linear entre a altura e peso de pokemons do mesmo tipo?**

**<p>Para ilustrar, usarei apenas um único tipo em nossa análise, o tipo GRAMA, que conta com Pokémons de diversos tamanhos e pesos.</p>**
 
<p>Para verificar se existe correlação linear entre duas variáveis, podemos plotar um gráfico de dispersão relacionando as variáveis que desejamos verificar.</p>

<p>Além disso, utilizamos também o Coeficiente de Pearson para verificar a correlação linear.</p>
<p>Coeficiente de Pearson:</p>
  
$$r_{xy} = \frac{\Sigma_{i}(x_{i}-\bar a)(y_{i}-\bar y)}{\sqrt{\Sigma_{i}(x_{i}-\bar x)^2(y_{i}-\bar y)^2}}$$

<p>Primeiramente criaremos tabelas de frequência com as variáveis que desejamos para cada tipo de Pokémon e então verificaremos a correlação linear com o Coeficiente de Pearson.</p>

```python
# creating a variable to print the number of unique values in the database
name_type_height_weight_df = df[['english_name', 'primary_type', 'height_m', 'weight_kg']]
display(name_type_height_weight_df)
```

```python
# filtering the dataframe to only show the grass type pokemon
grass_pokemon_filtering = np.where((name_type_height_weight_df['primary_type']=='grass'))

# defining a new dataframe with only the grass type pokemon
grass_pokemon_df = name_type_height_weight_df.loc[grass_pokemon_filtering]
display(grass_pokemon_df)
```

```python
# plotting a scatter plot to show the relationship between weight and height of the grass type pokemon
weight_kg_grass = grass_pokemon_df['weight_kg']
height_m_grass = grass_pokemon_df['height_m']

plt.scatter(weight_kg_grass, height_m_grass)
plt.title("Gráfico de Dispersão - Relação Peso/Altura Pokémon Tipo GRAMA - SEM MODELAGEM")
plt.xlabel("Peso - Kg")
plt.ylabel("Altura - M")
plt.show()

# using the .corr() method to print the Pearson correlation coefficient
corr_type_height = weight_kg_grass.corr(height_m_grass)
print(f'Este é o resultado do Coeficiente de Pearson:\n{corr_type_height}\n')
```

```python
# modeling the relationship between weight and height of the grass type pokemon - QUADRATIC MODEL
weight_kg_grass = grass_pokemon_df['weight_kg']
height_m_grass = grass_pokemon_df['height_m'] ** 2

plt.scatter(weight_kg_grass, height_m_grass)
plt.title("Gráfico de Dispersão - Relação Peso/Altura Pokémon Tipo GRAMA - MODELO QUADRÁTICO")
plt.xlabel("Peso - Kg")
plt.ylabel("Altura - M")
plt.show()

# using the .corr() method to print the Pearson correlation coefficient
corr_type_height_quad = weight_kg_grass.corr(height_m_grass)
print(f'Este é o resultado do Coeficiente de Pearson utilizando o Modelo Quadrático:\n{corr_type_height_quad}\n')
```

```python
# modeling the relationship between weight and height of the grass type pokemon - LOGARITMIC MODEL
weight_kg_grass = grass_pokemon_df['weight_kg'].apply(math.log)
height_m_grass = grass_pokemon_df['height_m']

plt.scatter(weight_kg_grass, height_m_grass)
plt.title("Gráfico de Dispersão - Relação Peso/Altura Pokémon Tipo GRAMA - MODELO LOGARITMICO LINEAR")
plt.xlabel("Peso - Kg")
plt.ylabel("Altura - M")
plt.show()

# using the .corr() method to print the Pearson correlation coefficient
corr_type_height_log = weight_kg_grass.corr(height_m_grass)
print(f'Este é o resultado do Coeficiente de Pearson utilizando o Modelo Logaritmico:\n{corr_type_height_log}\n')
```

## Conclusões iniciais:

### Modelagem:

<p>Podemos observar pelo Coeficiente de Pearson, sem nenhuma modelagem, que existe uma correlação linear direta forte, porém ao verificar o gráfico de dispersão é possível observar um comportamento de uma parábola nos dados puros, o que dificulta nossa análise, pois analisar uma reta é mais simples que analisar uma curva. Assim podemos verificar quais modelos de linearização, seriam mais adequados para uma previsão mais acurada.</p>

<p>Pode-se observar então que a relação entre as variáveis, é uma relação logaritmica, usaremos então o modelo Log-Linear, seu cálculo é mostrado abaixo:</p>

$$ y = \alpha + \beta \ln x $$

>**Pode-se perceber então que dentro da amostra (Pokémon tipo GRAMA) existe uma correlação linear forte entre a altura e peso.**

<p>Realizamos então a regressão linear</p>

```python
# defining the variables to be used in the model - LOGARITMIC MODEL
model = sm.OLS(height_m_grass, weight_kg_grass).fit()

print(f'Este é o resumo dos resultados que o modelo cálculou: \n {model.summary()} \n')
```

## Análise da Regressão Linear:

Podemos perceber pelo resultado acima, algumas informações interessantes como:

- o $r^{2}$: A variável dependente pode ser explicada por 89,9% da variável independente. Isso nos mostra uma correlação muito forte dentro da amostra.
- o $\beta$: A cada $0.3416$ temos a adição de 1 metro, porém como fizemos a modelagem na variável independente, temos que interpretar o $\beta$ como uma exponencial. Ou seja cada $e^{0.3416}$, temos o aumento de 1 m na altura.

```python
# plotting the linear regression model
plt.scatter(weight_kg_grass, height_m_grass)
plt.plot(weight_kg_grass, model.predict(weight_kg_grass), 'r-')
plt.title("Gráfico de Regressão Linear - Relação Peso/Altura Pokémon Tipo GRAMA")
plt.xlabel("Peso - Kg")
plt.ylabel("Altura - M")
plt.show()
```

```python
# predicting the height of a pokemon with the weight 
weight_kg = float(input("Insira o peso do Pokémon em kg: "))
predicted_height = model.predict(np.log(weight_kg))
print(f'A previsão de altura para um Pokémon com peso {weight_kg} kg é {predicted_height} m')

print('\nPOKEMON UMA ANÁLISE DE DADOS É UMA ANÁLISE INTERATIVA DOS DADOS ABERTOS DE POKEMONS ATÉ A VIII GERAÇÃO.\n')
print('Copyright (C) 2022  GABRIEL MACHADO')
print('This program is free software: you can redistribute it and/or modify it under the terms of the GNU General Public License as published by the Free Software Foundation,\neither version 3 of the License, or (at your option) any later version.\nThis program is distributed in the hope that it will be useful,\nbut WITHOUT ANY WARRANTY; without even the implied warranty of MERCHANTABILITY or FITNESS FOR A PARTICULAR PURPOSE. See the GNU General Public License for more details.\n\nYou should have received a copy of the GNU General Public License along with this program. If not, see http://www.gnu.org/licenses/')
```
