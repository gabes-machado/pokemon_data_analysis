### Perguntas-problema:

>**Existe correlação linear entre a altura e peso de pokemons do mesmo tipo?**
**<p>Para ilustrar, usarei apenas um único tipo em nossa análise, o tipo GRAMA, que conta com Pokémons de diversos tamanhos e pesos.</p>**

<p>Este é um estudo sobre técnicas de regressão linear e como podemos utilizá-la para resolver problemas em geral, que neste caso possui a temática de Pokémon. Dessa forma procurarei utilizar técnicas, e modelar matematicamente de forma a extrair o melhor desse tipo de análise.</p>

<p>Para verificar se existe correlação linear entre duas variáveis, vamos plotar um gráfico de dispersão relacionando as variáveis que desejamos verificar e utilizar um indicador de força para termos informações suficientes para realizar a análise.</p>

<p>Existem vários indicadores que podem ser utilizados para verificar a correlação entre duas variáveis, alguns dos mais comuns são:</p>
<ul>
    <li>Coeficiente de correlação de Pearson: Esse indicador varia de -1 a 1 e mede a força e a direção da relação linear entre as variáveis. Um valor próximo de 1 indica uma forte correlação positiva, enquanto um valor próximo de -1 indica uma forte correlação negativa;</li>
    <li>Coeficiente de correlação de Spearman: Esse indicador também varia de -1 a 1 e mede a força e a direção da relação não linear entre as variáveis. Ele é usado quando as variáveis não seguem uma distribuição normal;</li>
    <li>Coeficiente de correlação de Kendall: Esse indicador varia de -1 a 1 e mede a concordância entre as duas variáveis. Ele é usado quando as variáveis são ordinais ou categóricas;</li>
    <li>Entre diversos outros.</li>
</ul>

<p>Utilizaremos o Coeficiente de Pearson que é o mais comum para este tipo específico de relação. A equação de Pearson está indicada abaixo:</p>
<p>Coeficiente de Pearson:</p>
  
$$r_{xy} = \frac{\Sigma_{i}(x_{i}-\bar x)(y_{i}-\bar y)}{\sqrt{\Sigma_{i}(x_{i}-\bar x)^2(y_{i}-\bar y)^2}}$$

<p>Primeiramente criaremos tabelas de frequência com as variáveis que desejamos para cada tipo de Pokémon e então verificaremos a correlação linear com o Coeficiente de Pearson.</p>

```python
# Criação de um novo dataframe com as colunas 'english_name', 'primary_type', 'height_m' e 'weight_kg':
name_type_height_weight_df = df[['english_name', 'primary_type', 'height_m', 'weight_kg']]
display(name_type_height_weight_df)
```

```python
# Filtrando os dados para obter apenas os pokemons do tipo 'grass':
grass_pokemon_filtering = np.where((name_type_height_weight_df['primary_type']=='grass'))

# Definindo um novo dataframe com os dados filtrados:
grass_pokemon_df = name_type_height_weight_df.loc[grass_pokemon_filtering]
display(grass_pokemon_df)

plt.figure(figsize=(10,7), dpi= 80)
sns.displot(grass_pokemon_df['height_m'], color="dodgerblue", label="Compact",)
plt.show()
```

```python
# Plotando um gráfico de dispersão para verificar a relação entre o peso e a altura dos pokemons do tipo GRASS:
weight_kg_grass = grass_pokemon_df['weight_kg']
height_m_grass = grass_pokemon_df['height_m']

plt.scatter(weight_kg_grass, height_m_grass)
plt.title("Gráfico de Dispersão - Relação Peso/Altura Pokémon Tipo GRAMA - SEM MODELAGEM")
plt.xlabel("Peso - Kg")
plt.ylabel("Altura - M")
plt.show()

# Usando o método .corr() para verificar o coeficiente de correlação de Pearson:
corr_type_height = weight_kg_grass.corr(height_m_grass)
print(f'Este é o resultado do Coeficiente de Pearson:\n{corr_type_height}\n')
```

```python
# Modelando a relação entre o peso e a altura dos pokemons do tipo GRASS com um modelo quadrático:
weight_kg_grass = grass_pokemon_df['weight_kg']
height_m_grass_quad = grass_pokemon_df['height_m'] ** 2

plt.scatter(weight_kg_grass, height_m_grass_quad)
plt.title("Gráfico de Dispersão - Relação Peso/Altura Pokémon Tipo GRAMA - MODELO QUADRÁTICO")
plt.xlabel("Peso - Kg")
plt.ylabel("Altura - M")
plt.show()

# Usando o método .corr() para verificar o coeficiente de correlação de Pearson:
corr_type_height_quad = weight_kg_grass.corr(height_m_grass_quad)
print(f'Este é o resultado do Coeficiente de Pearson utilizando o Modelo Quadrático:\n{corr_type_height_quad}\n')
```

```python
# Modelando a relação entre o peso e a altura dos pokemons do tipo GRASS com uma função logaritmica:
weight_kg_grass_log = grass_pokemon_df['weight_kg'].apply(math.log)
height_m_grass = grass_pokemon_df['height_m']

plt.scatter(weight_kg_grass_log, height_m_grass)
plt.title("Gráfico de Dispersão - Relação Peso/Altura Pokémon Tipo GRAMA - MODELO LOGARITMICO LINEAR")
plt.xlabel("Peso - Kg")
plt.ylabel("Altura - M")
plt.show()

# Usando o método .corr() para verificar o coeficiente de correlação de Pearson:
corr_type_height_log = weight_kg_grass_log.corr(height_m_grass)
print(f'Este é o resultado do Coeficiente de Pearson utilizando o Modelo Logaritmico:\n{corr_type_height_log}\n')
```

### Conclusões iniciais:

### Modelagem:

<p>Podemos observar pelo Coeficiente de Pearson, sem nenhuma modelagem, que existe uma correlação linear direta forte, porém ao verificar o gráfico de dispersão é possível observar um curva logarítmica nos dados puros, o que por si só não atrapalha nossa análise. Nesses casos, seria possível utilizar um modelo de regressão de Poisson por exemplo, que considera esse comportamento logarítmico. Porém o objetivo do estudo é utilizar técnicas de regressão linear simples, assim verificamos quais modelos de linearização foram mais adequados para fornecer um resultado mais acurado do modelo preditivo.</p>

Pode-se observar então que a relação entre as variáveis, é uma relação logaritmica e que ao utlizamos a função para linearizarmos os dados, obtivemos um Coeficiente de Pearson superior, porém apenas essa análise não é suficiente. Observamos que pelo coeficiente de Pearson, o modelo quadrático é malhor, mas o gráfico não demonstra essa relação Dessa forma usaremos então o modelo Linear para verificar o $r^{2}$, e então verificar qual modelo melhor se encaixa. Com essas alterações temos as seguintes modificações nos modelos de regressão linear simples:

Modelo Log-Linear
$$y = \alpha + \beta \ln x$$

Modelo Linear
$$y = \alpha + \beta x$$

Modelo Quadrático-Linear
$$y^{2} = \alpha + \beta x$$

>**Pode-se perceber então que dentro da amostra (Pokémon tipo GRAMA) existe uma correlação linear forte entre a altura e peso.**

<p>Realizamos então a regressão linear</p>

```python
# Definindo o modelo de regressão linear:
model_no_modelling = sm.OLS(height_m_grass, weight_kg_grass).fit()

print(f'Este é o resumo dos resultados que o modelo cálculou: \n {model_no_modelling.summary()} \n')
```

```python
# Definindo o modelo de regressão linear com o modelo quadrático:
model_quad = sm.OLS(height_m_grass_quad, weight_kg_grass).fit()

print(f'Este é o resumo dos resultados que o modelo quadrático cálculou: \n {model_quad.summary()} \n')
```

```python
# Definindo o modelo de regressão linear com o modelo logaritmico:
model_log = sm.OLS(height_m_grass, weight_kg_grass_log).fit()

print(f'Este é o resumo dos resultados que o modelo logaritmico cálculou: \n {model_log.summary()} \n')
```

### Análise da Regressão Linear:

Podemos perceber pelo resultado acima, algumas informações interessantes como:

- o $r^{2}$: A variável independente pode ser explicada por 89,9% da variável dependente. Isso nos mostra uma correlação muito forte dentro da amostra.
- o $\beta$: A cada $0.3416$ temos a adição de 1 metro, porém como fizemos a modelagem na variável independente, temos que interpretar o $\beta$ como uma exponencial. Ou seja cada $e^{0.3416}$, temos o aumento de 1 m na altura.

```python
# Plotando o gráfico de regressão linear para o modelo logaritmico:
plt.figure(figsize=(10,7), dpi= 80)
plt.scatter(weight_kg_grass_log, height_m_grass)
plt.plot(weight_kg_grass_log, model_log.predict(weight_kg_grass_log), color='red')
plt.title("Gráfico de Regressão Linear - Relação Peso/Altura Pokémon Tipo GRAMA - MODELO LOGARITMICO LINEAR")
plt.xlabel("Peso - Kg")
plt.ylabel("Altura - M")
plt.show()
```

```python
# Previsão de altura para um pokemon qualquer: 
weight_kg = float(input("Insira o peso do Pokémon em kg: "))
predicted_height = model_log.predict(np.log(weight_kg))
print(f'A previsão de altura para um Pokémon com peso {weight_kg} kg é {predicted_height} m')

print('\nPOKEMON UMA ANÁLISE DE DADOS É UMA ANÁLISE INTERATIVA DOS DADOS ABERTOS DE POKEMONS ATÉ A VIII GERAÇÃO.\n')
print('Copyright (C) 2022  GABRIEL MACHADO')
print('This program is free software: you can redistribute it and/or modify it under the terms of the GNU General Public License as published by the Free Software Foundation,\neither version 3 of the License, or (at your option) any later version.\nThis program is distributed in the hope that it will be useful,\nbut WITHOUT ANY WARRANTY; without even the implied warranty of MERCHANTABILITY or FITNESS FOR A PARTICULAR PURPOSE. See the GNU General Public License for more details.\n\nYou should have received a copy of the GNU General Public License along with this program. If not, see http://www.gnu.org/licenses/')
```

Durante o estudo, pude também perceber que a concepção do modelo de regressão é basicamente ser um "chute enbasado", no sentido de o modelo utilizará dados passados, e com base neles caso haja correlação linear, estimar um valor que esteja dentro de uma margem de erro do modelo.</p>
